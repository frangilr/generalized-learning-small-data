```python
import pandas as pd
import os
```

```python
cats = ["bottle", "bowl", "cup", "door", "window", "chair", "table", "spoon"]
dfs = []
subjs = ["030FD", "031HW", "032SR", "033SE", "036MR", "037SM", "038RB", "039ED", "041EC",
          "042MB", "043MP", "045NA", "046TE", "047MS"]
for cat in cats:
    for subj in subjs:
        temp_df = pd.read_csv(f"../../data/PanelC/PE_graphs/subs/chunkedK6_label_shuffling_graphs_null_PE_10pct_{subj}_{cat}.csv")
        # print(temp_df.shape)
        dfs.append(temp_df)
        del temp_df

df = pd.concat(dfs, ignore_index=True)
print(df.shape)
```

```python
df.to_csv(f"../../data/PanelC/PE_graphs/chunkedK6_label_shuffling_graphs_null_PE_10pct.csv", index=False)
```
