```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path

folder_path = "runs"
combined_df = pd.concat([pd.read_csv(f) for f in Path(folder_path).glob("*.csv")], ignore_index=True)

combined_df["epoch"] = combined_df.groupby(["dataset", "seed"]).cumcount() + 1

# create epoch 0 with chance accuracy and prepend it
chance_df = combined_df[["dataset", "seed"]].drop_duplicates().copy()
chance_df["epoch"] = 0
chance_df["train_acc"] = 16.67
chance_df["test_acc"] = 16.67
combined_df = pd.concat([chance_df, combined_df], ignore_index=True)

sns.set_theme(style="whitegrid", palette="husl")

# train acc
plt.figure(figsize=(6, 5))
sns.lineplot(
    data=combined_df, x="epoch", y="train_acc", hue="dataset", linewidth=2
)
plt.title("Train Accuracy", fontweight='bold')
plt.axhline(16.67, color='red', linestyle='--', alpha=0.6, label='Chance')
plt.ylim(0, 100)
plt.legend().remove()
plt.tick_params(axis='both', labelsize=20)
sns.despine()
plt.tight_layout()
plt.savefig("train_accuracy.pdf", dpi=300, bbox_inches="tight")
# plt.close()
plt.show()

# test acc
plt.figure(figsize=(6, 5))
sns.lineplot(
    data=combined_df, x="epoch", y="test_acc", hue="dataset", linewidth=2
)
plt.title("Test Accuracy", fontweight='bold')
plt.axhline(16.67, color='red', linestyle='--', alpha=0.6, label='Chance')
plt.ylim(0, 100)
plt.legend().remove()
plt.tick_params(axis='both', labelsize=20) 
sns.despine()
plt.tight_layout()
plt.savefig("test_accuracy.pdf", dpi=300, bbox_inches="tight")
# plt.close()
plt.show()
```
